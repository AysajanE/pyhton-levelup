```python
import random

a = random.random() # generates a random float in the range of 0 to 1
print(a)
```

```python
# Continuous uniform
a = random.uniform(1, 10)
print(a)
```

```python
# Random integer between a and b
a = random.randint(1, 10)
print(a)
```

```python
# Normal random variate
a = random.normalvariate(0, 1)
print(a)
```

```python
mylist = list("ABCDEFGH")
print(mylist)
# Pick a random element from the list
a = random.choice(mylist)
print(a)
# If we want to pick multiple elements, we can use 'random.sample' and specify using second argument
# The elements picked are unique
a = random.sample(mylist, 3)
print(a)

# If we want to pick multiple elements that can be the same:
a = random.choices(mylist, k=3)
print(a)
```

```python
# Shuffle elements in a list
random.shuffle(mylist)
print(mylist)
```

```python
# Use seed to regenerate random numbers
# Random numbers generated by the same seed are always the same
random.seed(1)
print(random.random())
print(random.randint(1, 10))

random.seed(1)
print(random.random())
print(random.randint(1, 10))
```

```python
# For security applications such as generate random password, token, we can't use random since
# the numbers are reproducable. 
# In such situations, we should use 'secrets' package
# Reference: https://docs.python.org/3/library/secrets.html
import secrets

a = secrets.randbelow(10)
print(a)

a = secrets.randbits(4)
print(a)

mylist = list("ABCDEFGH")
a = secrets.choice(mylist)
print(a)
```

```python
# We can use numpy package to generate random numbers

import numpy as np

a = np.random.rand(3)
print(a)

a = np.random.rand(3, 3)
print(a)

a = np.random.randint(0, 10, 3) # excludes 10, last argument specifies size
print(a)

a = np.random.randint(0, 10, (3, 4))
print(a)
```

```python
# numpy shuffle
arr = np.array([[1,2,2],[4,5,6],[7,8,9]])
print(arr)
# Shuffle only shuffles along the first dimension
np.random.shuffle(arr)
print(arr)
```

```python
# We can also use seed to regenerate results

np.random.seed(1)
print(np.random.rand(3,3))

np.random.seed(1)
print(np.random.rand(3,3))
```
